# Data Science with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/home/phat06/Downloads/test_1/Amazon.csv')
df.head()
```

```python
df.info()
df['date']=pd.to_datetime(df['Date'])
```

```python
fig, ax = plt.subplots(1,2, figsize= (10,5))
sns.lineplot(data=df, x='Date',y='Open', ax = ax[0],color= 'r').set_title('open and close of Amazon stock')
sns.lineplot(data=df, x='Date',y='Close', ax = ax[1],color= 'b')
```
